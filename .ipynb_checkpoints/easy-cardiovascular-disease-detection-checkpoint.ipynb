### Deleting columns weight, height as we added bmi attribute derived from these attributes. This reduces dimensions our dataset and thus helps in future modeling. also convert age attribute from days to year unit. So dropping the age attribute.

```python
data.drop(['weight','height',"id","age"],axis=1,inplace=True)
```

### Encoding
# For cholesterol  we are encoding as 1: 'normal', 2: 'above normal', 3: 'well above normal'
# For gluc -> 1: 'normal', 2: 'above normal', 3: 'well above normal'

# For both of the above we can also apply one-hot encoding a well


```python
data['cholesterol']=data['cholesterol'].map({ 1: 'normal', 2: 'above normal', 3: 'well above normal'})
data['gluc']=data['gluc'].map({ 1: 'normal', 2: 'above normal', 3: 'well above normal'})
dummies = pd.get_dummies(data[['cholesterol','gluc']],drop_first=True)
data = pd.concat([data,dummies],axis=1)
data.drop(['cholesterol','gluc'],axis=1,inplace=True)
data.head()
```
